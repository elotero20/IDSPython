# Estilos y Color


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Estilos
Podemos establecer estilos particulares:

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

## Remover líneas

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

## Tamaño y apariencia

Podemos utilizar height y aspect para manipular los tamaños de nuestras visualizaciones

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',height=2,aspect=4,data=tips)
```

## Scale y Context

set_context() nos permite sobreescribir en parámetros establecidos por default:

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
