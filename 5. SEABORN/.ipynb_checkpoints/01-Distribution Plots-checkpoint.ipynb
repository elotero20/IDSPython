# Gráficas de Distribución

Estas son los tipos de gráfico que podemos utilizar si queremos realizar visualizaciónes de distribución

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

___
## Import Seaborn as sns

```python
import seaborn as sns
%matplotlib inline
```

## Data
Seaborn viene incluido con datasets de prueba

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## distplot

distplot muestra la distribución de un set univariable de observaciones

```python
sns.distplot(tips['total_bill'])
```

Para olvidarnos de la línea de tendencia y obtener un histograma usamos:

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jointplot

jointplot() nos permite mezclar dos distplots para observar datos de más de una variable. Tu decides qué tipo (**kind**) de gráfico vas a usar:
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## pairplot

Pariplot nos graficará todas las relaciones de gráficas en un dataset (para las columnas numéricas) y podemos separar por colores (para columnas categóricas)

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```
